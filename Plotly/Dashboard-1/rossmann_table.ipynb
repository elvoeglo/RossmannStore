```python
import plotly.express as px
import pandas as pd
# read csv with ";" as seperator
rsm = pd.read_csv("group_rossmann_dataprep.csv", sep=";")
```

```python
# show top rows
rsm.head()
```

```python
# show dtypes of columns
rsm.dtypes
```

```python
# Convert 'Date' to Date
rsm["Date"]=pd.to_datetime(rsm["Date"], format="%d.%m.%Y")
rsm.dtypes
```

```python
# Group By Date and StateName, aggregate by sum of Sales -> nyc.groupby (....).agg(....)
# check results
rsm_sales=rsm.groupby(["StateName", "Date"], as_index=False).agg({"Sales": "sum"})
```

```python
rsm_sales
```

```python
rsm_sales.dtypes
```

```python
rsm_sales
```

```python
# Create Line Chart of Date and Sales again
fig = px.line(rsm_sales, x="Date", y="Sales", color="StateName")
fig.update_layout(xaxis=dict(tickformat="%m-%Y"))
fig
```
